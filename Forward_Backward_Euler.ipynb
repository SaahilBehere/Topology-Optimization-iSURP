```python
# x(t) = e^2t
#dx/dt = f(x) = 2*x
import numpy as np
import matplotlib.pyplot as plt

dt = 0.1
t = np.arange(0,8,dt)
x_f = np.zeros(len(t))
x_f[0] = 1
x_b = np.zeros(len(t))
x_b[0] = 1

#f = lambda x : 2*x
func = lambda t: np.exp(2*t)
#x = func(t)

#forward euler
for i in range(len(t)-1):
    #x_f[i+1] = x_f[i] + dt*f(x_f[i])
    x_f[i+1] = x_f[i]*(1+2*dt)
    
    
```

```python
plt.plot(t,x_f,label = "forward")
#plt.plot(t,func(t), label = "function")
plt.legend()
```

```python
#backward euler
for i in range(len(t)-1):
    #x_b[-i-1] = x_b[-i] + dt*f(x_b[-i])
    x_b[i+1] = x_b[i]/(1-2*dt)
```

```python
plt.plot(t,x_b,label="backward")
#plt.plot(t,func(t), label = "function")
plt.legend()
```

```python
plt.plot(t,x_f,label="forward")
plt.plot(t,x_b,label="backward")
plt.plot(t,func(t), label = "function")
plt.legend()
```
